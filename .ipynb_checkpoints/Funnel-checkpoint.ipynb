# Funnel Analysis Data Challenge

The company CEO isn't very happy with the  volume of sales and, especially, of sales coming from new users. Therefore, she asked you to investigate whether there is something wrong in the conversion funnel or, in general, if you could suggest how conversion rate can be improved.

Specifically, she is interested in :
* A full picture of funnel conversion rate for both desktop and mobile
* Some insights on what the product team should focus on in order to improve conversion rate as well as anything you might discover that could help improve conversion rate.


## Insights

### Date

* At the start of March there is a drop in retention in the funnel.

### Device
* In March the retention from search to payment for Desktop users drops from 16% to 4% percent. 
* The number of sales in Jan/Feb to Mar/April for Desktop users dropped from 114 to 36.


* In March the retention from home to search for Mobile users drops from 80% to 20% percent.
* The number of sales over the 2 month period for Mobile users dropped from 248 to 54.


* The company is lost more sales from Mobile than Desktop.  
* Before the drop in engagement the retention at every stage in the funnel  is lower for Desktop users than Mobile users. 


* These changes in the funnel are statistically confirmed signifant through a chi squared test.

### Gender
* Gender does not contribute to the drop in engagement.


## Recommendations
An investigation to determine the drop in sales around March 1st is warranted.   Look for broken links or other changes to the platform.

* Start with the home to search page on the Mobile device as this accounts for the largest drop in sales.
* Then look at changes to the search to payment page on the Desktop platform. 

A further investigation to the overal performance of the Desktop platform is also warranted. It is bringing in half the number of sales as mobile despite having double the incomming traffic.

### Merge the Data Frames 
* On user_id 
* Rename page to correspond to the specific page (home / search / payment / confirmation).
* In new data frame visited pages are marked 1.  Unvisited pages are marked 0. 
* Change the format of date for future plotting

```python
df_home = df_home.rename(columns={'page': 'home'})
df_search = df_search.rename(columns={'page': 'search'})
df_payment = df_payment.rename(columns={'page': 'payment'})
df_confirmation = df_confirmation.rename(columns={'page': 'confirmation'})
dfs = [df_user, df_home, df_search, df_payment, df_confirmation]

df = reduce(lambda left,right: pd.merge(left,right,how = 'outer',on='user_id'), dfs)
df.loc[df['search'] ==  'search_page' , 'search'] = 1
df.loc[df['home'] ==  'home_page' , 'home'] = 1
df.loc[df['payment'] ==  'payment_page' , 'payment'] = 1
df.loc[df['confirmation'] ==  'payment_confirmation_page' , 'confirmation'] = 1
df = df.fillna(0)

df['date'] = df['date'].astype('datetime64[ns]')
df= df.set_index(pd.DatetimeIndex(df['date']))
df = df.sort_values('date')


df.head()
```

### Statistical Significance
In order to determine if these values are statisitically significant, a Chi-Squared test is run.   The pvalue from this test confirms that the probability of this drop in retention occuring by chance is 0.  


```python
#Desktop
df_temp = df.loc[df['device']=='Desktop']
df_temp = df_temp.resample('D').sum().fillna(0)
expected = df_temp.head(int(len(df_temp)/2)-1).sum()
observed = df_temp.tail(int(len(df_temp)/2)+1).sum()

#Create contingency table
a = [expected.payment, expected.search-expected.payment ]
b = [observed.payment, observed.search-observed.payment ]

chi2, pvalue, _ , _ = scipy.stats.chi2_contingency([a, b])
print(str.format('The p-value for the chi-squared test for the Desktop platform is {0:.2e}', pvalue))

#Mobile
df_temp = df.loc[df['device']=='Mobile']
df_temp = df_temp.resample('D').sum().fillna(0)
expected = df_temp.head(int(len(df_temp)/2)-1).sum()
observed = df_temp.tail(int(len(df_temp)/2)+1).sum()

#Create contingency table
a = [expected.search, expected.home-expected.search ]
b = [observed.search, observed.home-observed.search ]

chi2, pvalue, _ , _ = scipy.stats.chi2_contingency([a, b])

print(str.format('The p-value for the chi-squared test for the Mobile platform is {0:.2e}', pvalue))


```

## Funnel Visualization
A common way to visualize a funnel analysis is as a literal funnel!

```python
def plot_funnel(data):
    fig, axis = plt.subplots(figsize = (9,6))

    axis.plot(-data[::-1], range(4), lw = 3, color = 'blue', marker = 'o')
    axis.text(-data[0], 3, str(round(100.*data[0],1)), ha = 'right', va = 'top', color = 'red')
    plt.text(-data[1], 2, str(round(100.*data[1],1)), ha = 'right', va = 'top', color = 'red')
    plt.text(-data[2], 1, str(round(100.*data[2],1)), ha = 'right', va = 'top', color = 'red')
    plt.text(-data[3], 0, str(round(100.*data[3],1)), ha = 'right', va = 'top', color = 'red')

    axis.plot(data[::-1], range(4), lw = 3, color = 'blue', marker = 'o')
    axis.fill_between(data[::-1],range(4),3, alpha = 0.25, color = 'blue')
    axis.fill_between(-data[::-1],range(4),3, alpha = 0.25, color = 'blue')

        
    axis.set_ylabel('Page', size = 14)
    axis.set_yticks(range(4))
    axis.set_yticklabels(['Home','Search','Payment','Confirmation'][::-1], size = 12)
    
    
    [tick.label.set_fontsize(12) for tick in axis.xaxis.get_major_ticks()]
    axis.set_xlabel('Fraction retained at page', size = 14)
    
```

### The funnel with all data points

```python
counts = np.array([df.home.sum(), df.search.sum(), df.payment.sum(), df.confirmation.sum()])
counts_frac = counts/counts[0]
    
plot_funnel(counts_frac)
```

### Extended Visuals
This can be used to compare the funnels before and after March 1.  We can visualize the change in the funnel over time!

```python
# January and February
df_temp = df.resample('D').sum().fillna(0)
df_temp =  df_temp.head(int(len(df_temp)/2)-1).sum()

counts = np.array([df_temp.home.sum(), df_temp.search.sum(), df_temp.payment.sum(), df_temp.confirmation.sum()])
counts_frac = counts/counts[0]
print('\nFunnel for January and February') 
plot_funnel(counts_frac)
```

```python
# March and April
df_temp = df.resample('D').sum().fillna(0)
df_temp = df_temp.tail(int(len(df_temp)/2)+1).sum()
counts = np.array([df_temp.home.sum(), df_temp.search.sum(), df_temp.payment.sum(), df_temp.confirmation.sum()])
counts_frac = counts/counts[0]
print('\nFunnel for March and April') 
plot_funnel(counts_frac)
```
